## Задача 2

 $  f(x) = sin(\frac{x}{5})e^{\frac{x}{10}} + 5e^{\frac{-x}{2}}$

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1, 15, 0.01)
y = np.sin(x/5)*np.exp(x/10) + 5*np.exp(-x/2)

fig = plt.figure()
plt.plot(x, y)
plt.show()
```

Найдем значение функции в точках 1, 4, 8, 10, 15.

```python
def points(x):
    y = []
    for i in range(len(x)):
        y.append(np.sin(x[i]/5)*np.exp(x[i]/10) + 5*np.exp(-x[i]/2))
    return y
```

```python
x = [1.,4.,8.,10.,15.]
print(points(x))
```

Приближение многочленом первой степени вида $w_0 + w_1*x = f(x) $: По условию, рассмотрим функцию в точках 1 и 15.

$x + y = 3.2522168652714192
\\
x + 15*y = 0.63522141957866562 $

```python
a = np.array([[1,1],[1,15]])
x = [1.,15.]
b = points(x)
y = list(np.linalg.solve(a,b))
print('Answer', y)
#print(a.dot(y))
```

Приближение второго порядка многочленом  вида $w_0 +  w_1*x +w_2*x^2 = f(x)$

Аналогично:

$x + y + z = 3.2522168652714192
 \\
 x + y*8+ z*64 = 1.7468459495903677
 \\
 x + y*15 + z* 225 =  0.63522141957866562
$

```python
a = np.array([[1,1,1],[1,8,64],[1,15,225]])
x = [1.,8.,15.]
b = points(x)
y = list(np.linalg.solve(a,b))
print('Answer', y)
```

Для многочлена 3-го порядка в точках 1, 4, 10, 15:

```python
a = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000],[1,15,225,3375]])
x = [1.,4.,10.,15.]
b = points(x)
y = list(np.linalg.solve(a,b))
print('Answer', y)
```
